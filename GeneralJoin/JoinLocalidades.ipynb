```python
import geopandas as gpd
import os
import glob
import matplotlib.pyplot as plt
import geofeather as gf
import pandas as pd
```

```python
%%time
gfLocalidades = gf.from_geofeather('Localidades_4326.feather')
CRS_BASE = gfLocalidades.crs
```

```python
%%time
BASE_DIR = 'Coberturas'
strMaps = []
for (root, dirs, file) in os.walk(BASE_DIR):
    for f in file:
        if '.feather' in f.lower() and '.crs' not in f.lower():
            strMaps.append(os.path.join(root, f))
            
for i in range(len(strMaps)):
    print(i, strMaps[i])
```

```python
regionPath = strMaps
regionPath
```

```python
%%time
for region in regionPath:
    joinName = region.split('.feather')[0]
    print(joinName)
    
    #gdfjoinedv1 =  gdfloc.sjoin(gdfR1, how='left')
    
    %time gfCobertura = gf.from_geofeather(region)
    #%time gfCovered = gfLocalidades.sjoin(gfCobertura, how = 'left', rsuffix = 'Covered')
    %time gfCovered = gfLocalidades.sjoin(gfCobertura, how='left')
    #gfCobertura = gf.from_geofeather(region)
    #gfCovered = gfLocalidades.sjoin(gfCobertura, how = 'left', rsuffix = 'Covered')
    gfCovered.fillna(9999, inplace=True)
    
    print('Registros: ', gfCovered.shape)
    gpd.GeoDataFrame(gfCovered).to_csv(joinName + '.csv')
    print('###################################################################################################################')
```
