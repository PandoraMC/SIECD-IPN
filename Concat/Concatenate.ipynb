```python
ITER2010_Conditioned = ITER2010[c2010]
ITER2010_Conditioned.insert(loc = 17, column = 'Altán_4G_D', value = 0)
d = dict()
for i in range(len(c2020)):
    d[ITER2010_Conditioned.columns[i]] = c2020[i]
ITER2010_Conditioned.rename(d, axis = 'columns', inplace = True)
ITER2010_Conditioned['ITER'] = 'ITER 2010'
ITER2010_Conditioned['Trimestre'] = '3T 2010'
d
```

```python
ITER2020['ITER'] = 'ITER 2020'
```
